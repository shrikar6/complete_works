```python
from ortools.linear_solver import pywraplp
import time

solver = pywraplp.Solver.CreateSolver('SCIP')

bin_arr = []
for i in range(67):
    bin_arr.append([0]*5)

for i in range(len(bin_arr)):
    for j in range(len(bin_arr[0])):
        bin_arr[i][j] = solver.IntVar(0.0, 1.0, "{0},{1}".format(i, j))

for i in range(len(bin_arr)):
    solver.Add(sum(bin_arr[i]) == 1)
for j in range(len(bin_arr[0])):
    constraint_exp = 0
    for i in range(len(bin_arr)):
        constraint_exp += bin_arr[i][j]
    solver.Add(constraint_exp <= 14)
    solver.Add(constraint_exp >= 13)

matrix = [[9,4,9,7,7],
         [9,4,7,10,8],
         [7,8,9,7,10],
         [7,4,8,4,8],
         [10,7,6,0,5],
         [10,7,6,6,6],
         [7,6,9,7,8],
         [6,3,9,9,8],
         [7,5,7,5,10],
         [8,4,9,7,4],
         [7,7,7,3,9],
         [4,3,10,5,3],
         [6,5,9,6,6],
         [7,8,9,4,6],
         [10,8,5,6,8],
         [8,10,8,5,5],
         [6,8,7,5,9],
         [6,4,8,8,7],
         [7,6,5,9,9],
         [7,5,9,4,7],
         [8,6,8,7,8],
         [4,6,7,5,9],
         [5,3,10,9,4],
         [5,4,9,4,4],
         [7,6,10,4,7],
         [10,7,6,8,5],
         [5,5,10,6,5],
         [6,5,6,4,10],
         [7,5,10,8,3],
         [7,5,9,6,3],
         [7,6,8,7,3],
         [7,9,7,9,5],
         [8,5,7,7,7],
         [4,3,10,3,4],
         [6,5,10,6,6],
         [5,3,9,9,4],
         [5,4,6,10,4],
         [8,7,8,5,5],
         [9,8,6,5,5],
         [9,6,7,10,5],
         [7,5,8,9,5],
         [4,3,6,3,10],
         [4,4,10,7,7],
         [7,8,7,9,7],
         [10,4,7,10,6],
         [6,8,6,4,6],
         [7,6,8,4,7],
         [5,5,9,4,8],
         [5,5,10,7,6],
         [6,7,9,4,8],
         [5,1,10,4,9],
         [5,5,9,5,6],
         [7,5,5,6,10],
         [9,7,5,6,9],
         [6,5,8,4,8],
         [7,8,5,4,9],
         [8,7,6,7,10],
         [10,9,4,7,5],
         [3,5,10,3,5],
         [5,4,9,3,8],
         [9,9,3,6,4],
         [8,5,6,9,5],
         [7,8,10,7,6],
         [10,7,8,5,6],
         [10,6,8,4,4],
         [10,5,6,4,4],
         [8,4,7,8,9]]

obj_func = 0
for i in range(len(bin_arr)):
    for j in range(len(bin_arr[0])):
        obj_func += matrix[i][j]*bin_arr[i][j]
        
solver.Maximize(obj_func)
```
