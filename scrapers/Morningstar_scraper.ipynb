```python
with open("morningstar_etfs_risk_reward_info.pkl", "rb") as f:
    morningstar_etfs_risk_reward_info = pickle.load(f)
```

```python
scores = []
etfs = []

for etf in morningstar_etfs_risk_reward_info:
    score = 0
    for i in (3, 5, 10):
        if i in morningstar_etfs_risk_reward_info[etf]:
            risk, reward = morningstar_etfs_risk_reward_info[etf][i]
            if ((reward < 2 or risk > 2) and i == 3) or ((reward < 3 or risk > 2) and (i == 5 or i == 10)):
                score = -np.inf
                break
            else:
                score += reward-risk
    scores.append(score)
    etfs.append(etf)
```

```python
argsort = np.argsort(scores)[::-1]
for i in argsort:
    print(etfs[i] + ": " + str(morningstar_etfs_risk_reward_info[etfs[i]]))
    print("score: " + str(scores[i]) + "\n")
```
