```python
olddriver = webdriver.Firefox()
driver = webdriver.Firefox()
```

```python
olddriver.get("https://www.deckshop.pro/card/list")
```

```python
crdict = {}
```

```python
currcards = set(["Skeletons","Goblins","Spear Goblins","Bomber","Archers","Arrows","Knight","Minions","Cannon",
                 "Skeleton Dragons", "Barbarians","Tombstone","Mega Minion","Fireball","Mini P.E.K.K.A","Musketeer",
                "Goblin Cage","Valkyrie","Battle Ram","Giant","Goblin Barrel","Baby Dragon","Prince","Giant Skeleton",
                 "Inferno Tower","Fire Spirit","Wizard","P.E.K.K.A", "Skeleton Army","Bats","Zap","Hog Rider","Witch",
                 "Wall Breakers","Princess", "Mortar","X-Bow","Rocket","Flying Machine", "Balloon", "Royal Recruits",
                 "Royal Giant","Miner","Executioner","Mega Knight","Lightning","Freeze","Ice Spirit","Royal Hogs",
                 "Giant Snowball", "Three Musketeers","Dark Prince","Poison","Goblin Giant","Ice Golem","Goblin Gang",
                 "Goblin Hut","Hunter","Golem","Dart Goblin","Elite Barbarians","Minion Horde","Furnace"])
```

```python
images = olddriver.find_elements_by_tag_name("img")[:103]
hrefs = [a.get_attribute("href") for a in olddriver.find_elements_by_tag_name("a")[67:170]]
for i in range(103):
    card = images[i].get_attribute("alt")
    crdict[card] = set()
    driver.get(hrefs[i])
    try:
        atksyn = driver.find_element_by_id("atksyn")
        divs = atksyn.find_elements_by_tag_name("div")
        for div in divs:
            if div.get_attribute("class") != "opacity-25":
                crdict[card].add(div.find_element_by_tag_name("img").get_attribute("alt"))
    except NoSuchElementException:
        pass
    try:
        dfsyn = driver.find_element_by_id("dfsyn")
        divs = dfsyn.find_elements_by_tag_name("div")
        for div in divs:
            if div.get_attribute("class") != "opacity-25":
                crdict[card].add(div.find_element_by_tag_name("img").get_attribute("alt"))
    except NoSuchElementException:
        pass
```

```python
synlist = []
synpossible = crdict["Miner"].intersection(currcards)
for card in synpossible:
    synlist.append((len(synpossible.intersection(crdict[card])), card))
synlist.sort(reverse=True)
for i in synlist:
    print(i)
```

```python
currdeck = list(currcards)
newset = {}
for oldcard in currdeck:
    for card in crdict[oldcard]:
        if card not in newset:
            newset[card] = 0
        newset[card] += 1
for i in newset:
    if i in currcards:
        print(i, newset[i])
```
